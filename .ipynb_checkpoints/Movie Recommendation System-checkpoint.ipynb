```python
# Loading Libraries
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

# Load data from .dat files with ISO-8859-1 encoding
movies = pd.read_csv('movies.dat', sep='::', header=None, names=['movieId', 'title', 'genres'], engine='python', encoding='ISO-8859-1')
ratings = pd.read_csv('ratings.dat', sep='::', header=None, names=['userId', 'movieId', 'rating', 'timestamp'], engine='python', encoding='ISO-8859-1')
users = pd.read_csv('users.dat', sep='::', header=None, names=['userId', 'gender', 'age', 'occupation', 'zip'], engine='python', encoding='ISO-8859-1')

# Show the first few rows of each dataset
movies.head(), ratings.head(), users.head()
```

```python
# Merge datasets
movie_ratings = pd.merge(ratings, movies, on='movieId')

# Show merged data
movie_ratings.head()
```

```python
# Pivot the ratings data to create the user-item matrix
user_movie_ratings = movie_ratings.pivot_table(index='userId', columns='title', values='rating')

# Fill missing values with 0
user_movie_ratings = user_movie_ratings.fillna(0)

# Show a snippet of the user-item matrix
user_movie_ratings.head()
```

```python
# Calculate the similarity matrix
similarity_matrix = cosine_similarity(user_movie_ratings)

# Convert the similarity matrix to a DataFrame
similarity_df = pd.DataFrame(similarity_matrix, index=user_movie_ratings.index, columns=user_movie_ratings.index)

# Show a snippet of the similarity matrix
similarity_df.head()
```

```python
def recommend_movies(user_id, num_recommendations=5):
    # Get similar users sorted by similarity score
    similar_users = similarity_df[user_id].sort_values(ascending=False).index[1:num_recommendations+1]
    
    recommended_movies = []
    for user in similar_users:
        # Get movies rated by similar users
        movies_rated = user_movie_ratings.loc[user, user_movie_ratings.loc[user] > 0].index
        recommended_movies.extend(movies_rated)
    
    # Return top recommended movies, removing duplicates
    return list(set(recommended_movies))

# Example: Recommend movies for user 1
recommend_movies(user_id=1)
```

```python
# Plot the similarity matrix
plt.figure(figsize=(10, 8))
plt.imshow(similarity_matrix, cmap='coolwarm', interpolation='nearest')
plt.title('User Similarity Matrix')
plt.colorbar()
plt.show()
```

```python
def recommend_movies_weighted(user_id, num_recommendations=5):
    similar_users = similarity_df[user_id].sort_values(ascending=False).index[1:num_recommendations+1]
    
    movie_scores = {}
    for user in similar_users:
        for movie in user_movie_ratings.columns:
            if user_movie_ratings.loc[user, movie] > 0:  # Movie rated by user
                if movie not in movie_scores:
                    movie_scores[movie] = []
                movie_scores[movie].append(user_movie_ratings.loc[user, movie] * similarity_df[user_id][user])
    
    # Calculate weighted average scores
    weighted_avg_scores = {movie: np.mean(scores) for movie, scores in movie_scores.items()}
    
    # Sort movies by score
    recommended_movies = sorted(weighted_avg_scores.items(), key=lambda x: x[1], reverse=True)[:num_recommendations]
    
    return [movie[0] for movie in recommended_movies]

# Example: Recommend movies for user 1 with weighted ratings
recommend_movies_weighted(user_id=1)
```

```python
# Assuming you are recommending movies for user 1
user_id = 1  # Define user_id here

# Get the recommended movies
recommended_movies = recommend_movies(user_id=user_id)

# Limit the number of recommendations shown (optional, depending on the number of movies)
recommended_movies = recommended_movies[:10]  # Display top 10 recommendations (adjust as needed)

# Plot the recommended movies with horizontal bars
plt.figure(figsize=(12, 8))  # Increase figure size for better visibility
plt.barh(recommended_movies, range(len(recommended_movies)))

# Rotate labels to make them more readable
plt.xlabel('Recommended Movies')
plt.title(f'Movie Recommendations for User {user_id}')
plt.yticks(rotation=0)  # Rotate y-axis labels for better readability (you can adjust rotation if necessary)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# Suppose actual_ratings is the known ratings for the user
actual_ratings = ratings[ratings['userId'] == 1]  # Replace with your test set

# Get recommended movies and their ratings for the user (assuming prediction of ratings)
recommended_movies = recommend_movies(user_id=1)

# Get the movies from the 'movies' DataFrame and their corresponding ratings
recommended_movies_df = movies[movies['title'].isin(recommended_movies)]

# Merge with actual ratings to get the predicted ratings
# You need to consider only the movies the user has rated
predicted_ratings = actual_ratings[actual_ratings['movieId'].isin(recommended_movies_df['movieId'])]

# Now, actual_ratings['rating'] and predicted_ratings['rating'] should align
rmse = sqrt(mean_squared_error(predicted_ratings['rating'], [predicted_ratings['rating'].mean()] * len(predicted_ratings)))

print("RMSE:", rmse)
```
